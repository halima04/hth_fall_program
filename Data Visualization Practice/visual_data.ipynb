```python
import pandas
import seaborn 
import matplotlib.pyplot as plt
```

```python
# Call in our titanic csv
titanic_data = pandas.read_csv("titanic.csv")
titanic_data
```

```python
# Step 1: Create a heatmap showing the correlation between the following features: survival, pclass, sex, age, and fare 
titanic_features = titanic_data [["Survived", "Pclass", "Sex", "Age", "Fare"]]

seaborn.heatmap(titanic_features.corr(), annot = True, camp = 'mako')
plt.show
```

```python
# Step 2: Create a bar graph that illustrates how many passengers fell into each of these age ranges: 0 - 16, 17 - 25, 26 -40, 41 - 59, 60 or older
titanic_maxage = titanic_data ["Age"].max()
titanic_maxage

age_groups = {
      "groups": [
        "16 or younger",
        "17 to 25",
        "26 to 40",
        "41 to 60",
        "60+"

      ],

      "count": [
          len( titanic_data[ titanic_data ["Age"] <= 16]),
          len( titanic_data[ titanic_data ["Age"] < 16]) & len( titanic_data[ titanic_data ["Age"] <= 25]),
          len( titanic_data[ titanic_data ["Age"] < 25]) & len( titanic_data[ titanic_data ["Age"] <= 40]),
          len( titanic_data[ titanic_data ["Age"] < 40]) & len( titanic_data[ titanic_data ["Age"] <= 60]),
          len( titanic_data[ titanic_data ["Age"] <= 16])]

}

vis_df = pandas.DataFrame(age_groups)
seaborn.barplot(data = vis_df, x = 'groups', y = 'count');
plt.show()
```
